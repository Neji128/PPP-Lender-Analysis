## Step 1: Read raw PPP and location data from csv files

```python
import pandas as pd

# Reading PPP Dataset
PPP_df = pd.read_csv('../data/interim/PPP_concatenated.csv')
```

```python
PPP_df.head()
```

```python
# Reading Location Dataset
loc_df = pd.read_csv('../data/raw/loc_data/us_zip_fips_county.csv', encoding = "UTF-8")
```

```python
loc_df.head()
```

```python
PPP_df.columns
```

## Step 2: Create 5 digit zip codes for both datasets

```python
PPP_df["ZipCode"] = PPP_df['BorrowerZip'].str[:5]
```

```python
loc_df["ZipCode"]=loc_df['Zip Code'].apply(lambda x: '{:05d}'.format(x))
```

## Step 3: Merge location data to the PPP dataset

```python
pd.set_option('display.max_columns', None)
merged_df = PPP_df.merge(loc_df, how='left', on='ZipCode')
merged_df.head()
```

```python
merged_df.columns
```

```python
joined_df = PPP_df.join(loc_df, on='ZipCode', how='left', lsuffix='_t1', rsuffix='_t1')
joined_df.head()
```
