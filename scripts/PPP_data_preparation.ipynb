```python
df.head()
```

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
strings
```

```python
for col in strings:
    df[col] = df[col].str.upper()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Fill any remaining nulls with "Not Available" - only for string columns
for col in strings:
    df[col] = df[col].fillna('Not Available')
```

```python
df.to_csv('../data/cleaned/PPP_data/PPP_data.csv', encoding="utf-8")
```
