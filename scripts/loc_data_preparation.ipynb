# 1.0 Data preparation

```python
import numpy as np
import pandas as pd
import math
import base64
```

```python
# Read the data file and take a look at the data
df = pd.read_csv('../data/raw/loc_data/us_zip_fips_county.csv')
df.head()
```

```python
# Replace spaces in column names with underscores
df.columns = df.columns.str.replace(' ','_')
```

```python
df.head()
```

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
strings
```

```python
for col in strings:
    df_unique = pd.DataFrame(df[col].unique())
df_unique
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df_unique = pd.DataFrame(df_unique)
df_unique
```

```python
for col in strings:
    df[col] = df[col].str.upper()
```

```python
df.head()
```

```python
# Make Zip_Code 5 digits
df["Zip_Code"] = df['Zip_Code'].apply(lambda x: '{:05d}'.format(x))
```

```python
df.head()
```

```python
df.to_csv('../data/cleaned/loc_data/loc_data.csv')
```
